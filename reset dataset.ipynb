```python
import pandas as pd

print("🧬 正在启动数据融合协议 (Data Fusion)...\n")

# 1. 唤醒你的【保底存档】(包含所有旧数据)
old_df = pd.read_csv('wifi_data_v3.csv')
print(f"📦 旧存档样本总数: {len(old_df)}")

# 2. 唤醒你的【特种兵新存档】(你刚下楼扫的 5 个房间)
# 假设你把新收集的数据保存为了这个名字
new_df = pd.read_csv('wifi_data_new_5_rooms.csv')
print(f"🆕 新采集特种兵样本数: {len(new_df)}")

# 3. 列出那 5 个需要被“换血”的差生房间
bad_rooms = ['MSMR 3008', '2_Left_Stair', 'MSMR 3007', 'MSMX 0001', 'MSMX 2002']

# 4. 核心手术：从旧存档中，把这 5 个房间的数据【彻底剔除】！
# ~ 符号的意思是 "取反" (保留那些【不在】黑名单里的房间)
preserved_good_df = old_df[~old_df['location'].isin(bad_rooms)]
print(f"🛡️ 剔除差生后，保留的优等生样本数: {len(preserved_good_df)}")

# 5. 移花接木：把保留的优等生，和刚采集的特种兵合并！
final_v4_df = pd.concat([preserved_good_df, new_df], ignore_index=True)
print(f"🎉 融合完成！终极数据集样本总数: {len(final_v4_df)}")

# 6. 保存为全新的 V4 终极文件！
final_v4_df.to_csv('wifi_data_v4_final.csv', index=False)
print("\n💾 终极数据集已安全保存为 'wifi_data_v4_final.csv'！")
print("🚀 现在你可以用这个 v4 文件去跑 Random Forest 冲击 85% 以上了！")
```
